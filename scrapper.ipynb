# NIVM Dataset Scraper

```python
#Import libraries
from selenium.common.exceptions import StaleElementReferenceException
import time
import pandas as pd
import csv
import requests
import json
from selenium.webdriver.common.action_chains import ActionChains
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import chromedriver_autoinstaller
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium import webdriver
from bs4 import BeautifulSoup
```

```python
chromedriver_autoinstaller.install()  
driver = webdriver.Chrome()
url = "https://translate.google.com/translate?sl=auto&tl=en&u=https://lmm.rivm.nl/Tabel/Alle-jaren/Alle"
driver.get(url)
```

```python
driver.find_elements(By.CLASS_NAME, "hc_table_container")

all_dfs = []

# Scrape and click Next 287 times
for i in range(288):  
    
    containers = driver.find_elements(By.CLASS_NAME, "hc_table_container")
    for container in containers:
        html = container.get_attribute('outerHTML')
        soup = BeautifulSoup(html, "html.parser")
        table = soup.find("table")
        if table:
            df = pd.read_html(str(table))[0]
            all_dfs.append(df)

    try:
        next_btn = driver.find_element(By.ID, "btnNext")  
        next_btn.click()
        time.sleep(1)
    except:
        print("Next button not found")
        break

# Combine and save
if all_dfs:
    combined_df = pd.concat(all_dfs, ignore_index=True)
    combined_df.to_csv("nivm_dataset.csv", index=False)

driver.quit()
```
